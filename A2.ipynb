````python
# data loader function from assignment 1
def get_data_loader(data_dir, batch_size, valid_split=0.2):
    """
    Load the Mechanical Tools Dataset, split into training, validation, and testing.

    Args:
        data_dir: The root directory of the dataset, with subdirectories for each class.
        batch_size: Number of samples per batch.
        valid_split: Fraction of the training data to be used for validation.

    Returns:
        train_loader: Iterable DataLoader for training data.
        val_loader: Iterable DataLoader for validation data.
        test_loader: Iterable DataLoader for testing data.
        classes: List of class names.
    """
    # Read


    # Define transformations
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),         # Convert to Tensor
        transforms.Normalize((0.5,), (0.5,))  # Normalize to [-1, 1]
    ])

    # Load the full dataset
    dataset = datasets.ImageFolder(data_dir, transform=transform)
    classes = dataset.classes  # Get class names

    #training, validation, testing proportion
    test_person_id = list(range(91,103))
    train_person_id = list(range(1,73))
    val_person_id = list(range(73,91))

    train_index = []
    val_index = []
    test_index = []

    # Sort the data according to the person_id
    for idx, (path, label) in enumerate(dataset.samples):
      person_id = path.split('/')[-1].split("_")[0]

      if int(person_id) in train_person_id:
        train_index.append(idx)
      elif int(person_id) in val_person_id:
        val_index.append(idx)
      elif int(person_id) in test_person_id:
        test_index.append(idx)

    # Define the dataloader
    train_dataset = torch.utils.data.Subset(dataset, train_index)
    val_dataset = torch.utils.data.Subset(dataset, val_index)
    test_dataset = torch.utils.data.Subset(dataset, test_index)

    # Transform to dataloader
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

    return train_loader, val_loader, test_loader, classes
````

````python
# TO BE COMPLETED

# Define the model (3 Convlutional Layer + 2 Fully Connected Layer)
class MNISTClassifier(nn.Module):
  def __init__(self, conv_channels=[16,32,64], fc_hidden_units=128):
    super(MNISTClassifier, self).__init__()

    self.name=f"CNN_conv({conv_channels[0]}_{conv_channels[1]}_{conv_channels[2]})_fcunit{fc_hidden_units}"
    self.conv_channels = conv_channels

    self.conv1 = nn.Conv2d(
        in_channels=3,
        out_channels=conv_channels[0],
        kernel_size=3,
        padding = 1 #This preserve the input shape
    )
    self.conv2 = nn.Conv2d(
        in_channels=conv_channels[0],
        out_channels=conv_channels[1],
        kernel_size=3,
        padding = 1 #This preserves the input shape
    )
    self.conv3 = nn.Conv2d(
        in_channels=conv_channels[1],
        out_channels=conv_channels[2],
        kernel_size=3,
        padding = 1 #This preserves the input shape
    )
    self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    self.fc1 = nn.Linear(conv_channels[2]*28*28, fc_hidden_units)
    self.fc2 = nn.Linear(fc_hidden_units, 9)

  def forward(self, x):
    x = self.pool(F.relu(self.conv1(x)))
    x = self.pool(F.relu(self.conv2(x)))
    x = self.pool(F.relu(self.conv3(x)))
    x = x.view(-1, self.conv_channels[2]*28*28)
    x = F.relu(self.fc1(x))
    x = self.fc2(x)
    return x
````

Training Pipeline

````python
# Define the training pipeline
def train_CNN(model, data_dir, batch_size=64, n_epochs = 5, learning_rate=0.001, data_loader = get_data_loader):

  train_loader, val_loader, test_loader, classes = data_loader(data_dir, batch_size)
  criterion = nn.CrossEntropyLoss()

  # SGD or Adam
  # optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9)
  optimizer = optim.Adam(model.parameters(), lr=learning_rate)

  # Record the loss and accuracy in each epochs
  iters, train_losses, val_losses, train_error, val_error = [],[],[],[],[]

  # Check which device is used
  if use_Cuda and torch.cuda.is_available():
    print("GPU is used")
    model.cuda()

  # Trainig pipeline
  n = 0 # Number of iterations
  for epoch in range(n_epochs):
    for imgs, labels in train_loader:
      model.train()
      # Enablue GPU Usage
      if use_Cuda and torch.cuda.is_available():
        imgs = imgs.cuda()
        labels = labels.cuda()

      n += 1
      # Forward Pass
      out = model(imgs)
      loss = criterion(out, labels) # Calculate loss

      # Backward Pass
      loss.backward() # Calculate gradients
      optimizer.step() # Update weights
      optimizer.zero_grad() # Reset the gradients

      # Save the current training information
      iters.append(n)

      train_err, train_loss = evaluate(model, criterion, train_loader)
      val_err, val_loss = evaluate(model, criterion, val_loader)

      train_error.append(train_err)
      val_error.append(val_err)
      train_losses.append(train_loss)
      val_losses.append(val_loss)

      if n % 10 == 0:
        print(f"Iteration {n}: Train err: {train_error[-1]:.4f}, Train loss: {train_losses[-1]:.4f} | "
                f"Validation err: {val_error[-1]:.4f}, Val loss: {val_losses[-1]:.4f}")

      # Checkpoint: Save the current model state
      model_path = get_model_name(model.name, batch_size, learning_rate, n)
      torch.save(model.state_dict(), model_path)

  # Return the iteration that has the best performance
  best_iteration = np.argmin(val_error)
  print(f"Best Performance Iteration: {best_iteration+1}, Validation Error: {val_error[best_iteration]}")

  # plotting
  plt.figure()
  plt.title("Train vs Validation Error")
  plt.plot(range(1, len(train_error) + 1), train_error, label="Train")
  plt.plot(range(1, len(val_error) + 1), val_error, label="Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Error")
  plt.legend(loc='best')
  plt.show()

  plt.figure()
  plt.title("Train vs Validation Loss")
  plt.plot(range(1, len(train_losses) + 1), train_losses, label="Train")
  plt.plot(range(1, len(val_losses) + 1), val_losses, label="Validation")
  plt.xlabel("Epoch")
  plt.ylabel("Loss")
  plt.legend(loc='best')
  plt.show()
````

## Part 4. Transfer Learning [8pt]
For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an
ImageNet pre-trained AlexNet model to extract features in this assignment.

### Part (i) - [1pt EXPLORATORY]
Here is the code to load the AlexNet network, with pretrained weights. When you first run the code, PyTorch
will download the pretrained weights from the internet.

````python
import torchvision.models
alexnet = torchvision.models.alexnet(pretrained=True)
````

The alexnet model is split up into two components: *alexnet.features* and *alexnet.classifier*. The
first neural network component, *alexnet.features*, is used to compute convolutional features, which are
taken as input in *alexnet.classifier*.

The neural network alexnet.features expects an image tensor of shape Nx3x224x224 as input and it will
output a tensor of shape Nx256x6x6 . (N = batch size).

Compute the AlexNet features for each of your training, validation, and test data. Here is an example code
snippet showing how you can compute the AlexNet features for some images (your actual code might be
different):

**Save the computed features**. You will be using these features as input to your neural network in Part
(ii), and you do not want to re-compute the features every time. Instead, run *alexnet.features* once for
each image, and save the result.

````python
def extract_alexnet_features(alexnet, data_loader):
  alexnet.eval()
  features = []
  labels = []

  with torch.no_grad():
    for imgs, lbls in data_loader:
      if use_Cuda and torch.cuda.is_available():
        imgs = imgs.cuda()
        alexnet.cuda()
      feature = alexnet.features(imgs)
      features.append(feature.cpu())
      labels.append(lbls)

  # Concatenate all features and labels into single tensors
  features_tensor = torch.cat(features, dim=0)
  labels_tensor = torch.cat(labels, dim=0)

  # Create new dataset and dataloader
  dataset = TensorDataset(features_tensor, labels_tensor)
  dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

  return dataloader
````

Data Preprocessing

````python
# Create one single dataloader

data_dir = "/content/A2_Hand_Gesture_Unlabeled_Data/"

batch_size = 1

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to 128x128
    transforms.ToTensor(),         # Convert to Tensor
    transforms.Normalize((0.5,), (0.5,))  # Normalize to [-1, 1]
])

# Load the full dataset
dataset = datasets.ImageFolder(data_dir, transform=transform)

# Sort data by filename
dataset.samples.sort(key=lambda x: int(x[0].split('/')[-1].split('.')[0]))  # Sort by file path

# Create one single dataloder for the data
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False)
````
